```python

#! 简答题:
#todo 1、回归任务和分类任务的联系与区别。
# 简单来讲，分类任务 和 回归任务 的区别在于 需要预测的值的类型：
# 回归任务，是对 连续值 进行预测（比如 多少）；
# 分类任务，是对 离散值 进行预测（比如 是不是，属不属于，或者 属于哪一类）。
#* 即输入变量与输出变量均为连续变量的预测问题是回归问题,输出变量为有限个离散变量的预测问题成为分类问题。

#todo 2、线性回归和逻辑斯蒂回归的关系。
# 1）线性回归要求变量服从正态分布，logistic回归对变量分布没有要求。
# 2）线性回归要求因变量是连续性数值变量，而logistic回归要求因变量是分类型变量。
# 3）线性回归要求自变量和因变量呈线性关系，而logistic回归不要求自变量和因变量呈线性关系
# 4）logistic回归是分析因变量取某个值的概率与自变量的关系，而线性回归是直接分析因变量与自变量的关系

#回归任务 因为:是对 连续值 进行预测
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
data=pd.read_csv('boston.csv')
data.head()#! 默认读取前五行的数据
```

```python
# dataframe 的操作函数
#!https://wenku.baidu.com/view/708389558d9951e79b89680203d8ce2f01666542.html
column_name=data.columns#! 注意columns后面不加括号
print(list(column_name))
column_name=list(column_name)#! 注意 medv 即为房价值
```

```python

#! 使用前400条作为训练集
Xs=data.drop(['medv'],axis=1)
train_x=Xs[:400]
print(Xs)
train_y=data['medv'][:400].to_numpy().reshape(-1,1)#! 注意这个values后面不加括号
y=data['medv'].values.reshape(-1,1)
print(y)

reg=LinearRegression()
# reg.fit(test_x,test_y)
reg.fit(train_x,train_y)
print(reg.coef_)
print(reg.intercept_)
print("The linear model is: Y = {:.5} + {:.5}*crim + {:.5}*zn + {:.5}*indus + {:.5}*chas + {:.5}*nox + {:.5}*rm + {:.5}*age + {:.5}*dis + {:.5}*rad + {:.5}*tax + {:.5}*ptratio + {:.5}*black + {:.5}*lstat ".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1], reg.coef_[0][2],reg.coef_[0][3], reg.coef_[0][4], reg.coef_[0][5],reg.coef_[0][6], reg.coef_[0][7], reg.coef_[0][8],reg.coef_[0][9], reg.coef_[0][10], reg.coef_[0][11],reg.coef_[0][12]))

test_x=Xs[400:]
test_y=data['medv'][400:].to_numpy().reshape(-1,1)


#! 用R^2判断模型拟合效果
print('R^2=',reg.score(test_x,test_y))#作用：返回该次预测的系数R2 
  
#todo 选择R²，R²是指拟合优度，是回归直线对观测值的拟合程度。
 
```

```python
from sklearn.decomposition import PCA
import numpy as np

score=0
best_x=0
for x in range(1,14):
    
    pca=PCA(n_components=x)
    pca.fit(Xs)
    X=pca.transform(Xs)
    reg.fit(X,y)
    if score<=reg.score(X,y):
        best_x=x
        score=reg.score(X,y)
        print(best_x,score)
        
print('===============')   
print(best_x,score)
#! 我们发现 随着维数的升高 拟合效果越来越好
    
```
